```python
# NOT KEEPING THIS
# budget
# homepage
# id
# original_language
# original_title
# popularity
# production_comapny
# production_countries
# release-date(not sure)
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]   # Only keeping these
movies.head(1)
```

```python
#Pre-Processing the data
movies.isnull().sum()  # Checking any null values
```

```python
movies.dropna(inplace=True)  # Dropping the null values
movies.isnull().sum()
```

```python
movies.duplicated().sum()    #  Checking any duplicate values
```

```python
movies.iloc[0].genres
```

```python
# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
# ['Action','Adventure','Fantasy','science fiction']   --->  We need this form

import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):   # Converts string to a list
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)    # Changing the genres column
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)   # Changing the keywords column
movies.head()
```

```python
def convert3(text):     # To get name of first 3 cast only
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L

movies['cast'] = movies['cast'].apply(convert)   # Changing the cast column
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':     # Only takes the director values
            L.append(i['name'])
    return L
movies['crew'] = movies['crew'].apply(fetch_director)   # Changing the crew co
```

```python
 movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())   # Changes the overview column from string to a list
movies.sample(5)
```

```python
# To resolve any space between elements in th list
# [Edward Norton    ---->     EdwardNorton]
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# Changing the columns using collapse()
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
# creating tags column which consists of all the following columns together
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])   # Dropping these columns
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))   # Joining the tags column from list to a string
```

```python
new.head()
```

```python
new['tags'] =new['tags'].apply(lambda x:x.lower())   # To convert tags column into lower case
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()     # Used to stem the data in tags column
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new['tags'] =new['tags'].apply(stem)
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')  # Vectorizes the text and also remove the stop words
```

```python
vector = cv.fit_transform(new['tags']).toarray()   # Converting to a sparse array
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)    # Distance of a movie with all others (4806,4806)
```

```python
similarity.shape
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]     # To get the index of the movie
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])  # Movies are sorted according to similarity but mintaining the indexes
    for i in distances[1:10]:
        print(new.iloc[i[0]].title)   # Returns the title of first 5 similar movies
```

```python
recommend('Gandhi')
```
